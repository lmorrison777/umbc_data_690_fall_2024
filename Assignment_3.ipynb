```python
import pandas as pd
import plotly.express as px
data_url = "https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/fall2024/survey/UMBC%20DATA%20690%20FALL%202024%20DR%20WANG%20Student%20Survey.csv"

df = pd.read_csv(data_url)

df.head()
```

# Step One - Load the Survey Response Data and Explore

```python
df = pd.read_csv(data_url)

df.shape
```

# Step Two - Explore Student Age

```python
yob_list = list(df["8. My Birth Year"])
yob_list

cleansed_yob_list = []


for yob in yob_list:
    cleansed_yob = yob[-4:]
    cleansed_yob_list.append(cleansed_yob)

cleansed_yob_list
```

```python
age_list = []

for yob in cleansed_yob_list:
    age = 2024 - int(yob)
    age_list.append(age)

age_list
```

# Step 3 - Explore Educational Background

```python
edu_list = list(df["1. My primary educational background"])

edu_list



```

```python
df_edu_dist = df["1. My primary educational background"].value_counts()
df_edu_dist
```

```python
df_edu_dist = df_edu_dist.reset_index()
df_edu_dist
```

```python
fig = px.bar(
    df_edu_dist,
    x="1. My primary educational background",
    y="count",
    color="1. My primary educational background",
    title="UMBC DATA 690 Fall 2024 Student Education Distribution"
)

fig.show()
```

##Homework: Create a pie chart using plotly express

```python
import plotly.express as px
edu_dictionary = dict(df["1. My primary educational background"].value_counts())
edu_dictionary

```

```python
labels = edu_dictionary.keys()
amount = edu_dictionary.values()
fig = px.pie(names = labels, values = amount, title = 'Primary Educational Background')
fig.show()
```
