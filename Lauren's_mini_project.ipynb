## Step 5 - Aggregate Data

Aggregate the following columns at the "Group" level for each year:

- For GDP, using "sum" for aggregation.
- For GDP Per Capita, use "mean" for aggregation.
- Check your calculation with the reference site.

```python
#GDP
gdp_sum_per_year = df.groupby(['Year', 'Group'])['GDP ($Billions)'].sum().reset_index()
gdp_sum_per_year
```

```python
per_capita_gdp_mean = df.groupby(['Year', 'Group'])['GDP Per Capita ($)'].mean().reset_index()
per_capita_gdp_mean
```

## Step 6 - Calculate Percentage Share of GDP

For each year, calculate the percentage share of each group's GDP. Check your calculation with the reference site.

```python
pd.set_option('display.max_columns', None)
total_gdp_per_year = df.groupby('Year')['GDP ($Billions)'].sum().reset_index()
total_gdp_per_year.rename(columns={'GDP ($Billions)': 'Worldwide GDP ($Billions)'}, inplace=True)
total_gdp_per_year
```

```python
df = pd.merge(df, total_gdp_per_year, on='Year', suffixes=('', '_Total'))
df.sample(5)
```

```python
gdp_by_region_year = df.groupby(['Year', 'Group'])['GDP ($Billions)'].sum().reset_index()
gdp_by_region_year = pd.merge(gdp_by_region_year, df[['Year', 'Worldwide GDP ($Billions)']].drop_duplicates(), on='Year')
gdp_by_region_year['Percent Share'] = (gdp_by_region_year['GDP ($Billions)'] / gdp_by_region_year['Worldwide GDP ($Billions)']) * 100
gdp_by_region_year
```

## Step 7 - Plot the GDP Percentage Share Over Time

```python
fig = px.line(gdp_by_region_year, x='Year',
              y='Percent Share',
              color='Group',
              title='Percentage Share of GDP by Group Over Time')

fig.show()
```
