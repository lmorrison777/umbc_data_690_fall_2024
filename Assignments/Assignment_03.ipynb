##Assignment 03 Homework Part 2
###Process column "11. What is your long-term career goal?" to find out the most popular career goals of the class. User Python code to parse the responses and count occurences of words, then use your naked eyes to make determination. Write your observation/analysis using Markdown.

```python
careers = list(df["11. What is your long-term career goal?"])
careers

parsed_careers = {}

for career in careers:
  words = career.split()
  for word in words:
    if word in parsed_careers:
      parsed_careers[word] += 1
    else:
      parsed_careers[word] = 1

parsed_careers
```

##Analysis:
- Some of the most common significant words were: analyst, data, career, work, and problems.
- Based on this, it appears that working/having a career is a long term goal (meaning people are taking this class to learn skills to advance their ability to work, not just taking it for fun)
- Students want to be able to work with data, including in careers such as analysts
